```python
import pandas as pd
from  matplotlib import pyplot as plt
plt.rcdefaults()
plt.style.use('fivethirtyeight')

data=pd.read_csv('covid19_recovered_southAsia.csv')
daycount=data['Days']


BD=data['Bangladesh']
SL=data['Sri Lanka']
AG=data['Afghanistan']
ID=data['India']
NL=data['Nepal']
BT=data['Bhutan']
PK=data['Pakistan']
ML=data['Maldives']

plt.plot(daycount,BD,Label="Bangladesh")
plt.plot(daycount,SL,Label="Sri Lanka")
plt.plot(daycount,AG,Label="Afghanistan")
plt.plot(daycount,ID,Label="Nepal")
plt.plot(daycount,BT,Label="Bhutan")
plt.plot(daycount,PK,Label="Pakistan")
plt.plot(daycount,ML,Label="Maldives")


plt.legend()
plt.title('covid 19 Recovering Analysis in South Asia')
plt.ylabel('Number of Recoveries')
plt.xlabel('Number of Days')
plt.savefig("plot.png")
plt.xlim(0,500)
plt.ylim(0,100000)
plt.show()










```
